#### IMPORT FROM DATABASES

```python
from sqlalchemy import create_engine, inspect
```

```python
db_file = 'sqlite:///datasets/data.db'
```

```python
engine = create_engine(db_file)
```

```python

inspector = inspect(engine)
print(inspector.get_table_names())
```

```python
hpd_calls = pd.read_sql('hpd311calls', engine)

# View the first few rows of data
print(hpd_calls.head())
```

```python
query = """
SELECT * 
  FROM weather;
"""

# Load weather with the SQL query
weather = pd.read_sql(query, engine)

# View the first few rows of data
print(weather.head())
```

```python
query2 = """
select *
from hpd311calls
where complaint_type ='SAFETY';
"""

# Query the database and assign result to safety_calls
safety_calls = pd.read_sql(query2,engine)

# Graph the number of safety calls by borough
call_counts = safety_calls.groupby('borough').unique_key.count()
call_counts.plot.barh()
plt.show()
```

```python
query3 = """
SELECT *
  FROM weather
  where tmax <= 32 OR snow >= 1 
 ;
"""

# Query database and assign result to wintry_days
wintry_days = pd.read_sql(query3, engine)

# View summary stats about the temperatures
print(wintry_days.describe())
```

```python
calls_by_issue_query = """
SELECT complaint_type, 
       COUNT(*) AS call_count
  FROM hpd311calls
 GROUP BY complaint_type
 ORDER BY call_count DESC;
"""

# Create dataframe of call counts by issue
calls_by_issue = pd.read_sql(calls_by_issue_query, engine)

# Sort ascending so largest bar appears at the top of a horizontal chart
calls_by_issue = calls_by_issue.sort_values("call_count", ascending=True)

# Plot
ax = calls_by_issue.plot.barh(x="complaint_type", y="call_count", legend=False, figsize=(8, 6))
ax.set_xlabel("Number of calls")
ax.set_ylabel("Complaint type")
ax.set_title("HPD 311 calls by complaint type")

# Add count labels at the end of each bar
ax.bar_label(ax.containers[0], padding=3)

# Give the labels a bit of breathing room so they aren't clipped
ax.margins(x=0.1)

plt.tight_layout()
plt.show()
```

```python
month_query = """
SELECT month, 
       MAX(tmax),
       MIN(tmin),
       SUM(prcp)
  FROM weather 
 GROUP BY month;"""

# Get dataframe of monthly weather stats
weather_by_month = pd.read_sql(month_query, engine)

# View weather stats by month
print(weather_by_month)
```

```python
query = """
SELECT hpd311calls.*, weather.prcp
  FROM hpd311calls
  JOIN weather
    ON hpd311calls.created_date = weather.date
WHERE hpd311calls.complaint_type = 'WATER LEAK';"""

# Load query results into the leak_calls dataframe
leak_calls = pd.read_sql(query, engine)

# View the dataframe
print(leak_calls.head())
```

```python
query = """
SELECT hpd311calls.created_date, 
       COUNT(*)
  FROM hpd311calls 
 WHERE hpd311calls.complaint_type = 'HEAT/HOT WATER' 
 GROUP BY hpd311calls.created_date;
"""

# Query database and save results as df
df =  pd.read_sql(query, engine)

# View first 5 records
print(df.head())
```

```python
# Modify query to join tmax and tmin from weather by date
query = """
SELECT hpd311calls.created_date, 
	   COUNT(*), 
       weather.tmax,
       weather.tmin
  FROM hpd311calls 
       join weather
       ON hpd311calls.created_date = weather.date
 WHERE hpd311calls.complaint_type = 'HEAT/HOT WATER' 
 GROUP BY hpd311calls.created_date;
 """

# Query database and save results as df
df = pd.read_sql(query, engine)

# View first 5 records
print(df.head())
```

#### JSON FILES

```python
json_file = 'datasets/dhs-daily-report.json'
```

```python
pop_in_shelters = pd.read_json(json_file)

print(pop_in_shelters.describe())
```

```python
try:
    df = pd.read_json(json_file)
    
    normalized_columns = {col.strip().lower().replace(" ", "_"): col for col in df.columns}
    date_col = normalized_columns.get("date_of_census")
    y_col = normalized_columns.get("total_individuals_in_shelter")

    if date_col is None or y_col is None:
        print("Available columns:", list(df.columns))
        raise KeyError("required column not found")
    
    df[date_col] = pd.to_datetime(df[date_col])
    df.plot(x=date_col, y=y_col)
    plt.show()

except ValueError:
    print("pandas could not parse the JSON.")
except KeyError as exc:
    print("Required column missing:", exc)
```
